** Create a function that counts the number of times the word "dog" occurs in a string. Again ignore edge cases. **

```python
def countDog(s):
    A=s.split()
    count=0
    for word in A:
        if word=='dog':
            count+=1
    return count
```

**You are driving a little too fast, and a police officer stops you. Write a function
  to return one of 3 possible results: "No ticket", "Small ticket", or "Big Ticket". 
  If your speed is 60 or less, the result is "No Ticket". If speed is between 61 
  and 80 inclusive, the result is "Small Ticket". If speed is 81 or more, the result is "Big    Ticket". Unless it is your birthday (encoded as a boolean value in the parameters of the function) -- on your birthday, your speed can be 5 higher in all 
  cases. **

```python
def caught_speeding(speed, is_birthday):
    s=0
    if is_birthday:
        s=s+5
        
    if speed>=81+s:
        return 'Big Ticket'
    else:
        if speed >=61+s:
            return 'Small Ticket'
        else:
            return 'No Ticket'
```

```python
caught_speeding(65,True) #Check your function with different values
```

Write a function to find factorial of a number n. Factorial '!' is defined as:
n! = n x (n-1) x (n-2) . ...... 2 x 1

For example factorial of 7 is:

7! = 7 x 6 x 5 x 4 x 3 x 2 x 1 = 5040

```python
def factorial(n):
    a=n
    while (a>1):
        n=n*(a-1)
        a=a-1
    return n
```

```python
print(factorial(7)) #Call the function
```

```python
print(findWord(["P>E","E>R","R>U"])) # PERU
print(findWord(["I>N","A>I","P>A","S>P"])) # SPAIN
print(findWord(["U>N", "G>A", "R>Y", "H>U", "N>G", "A>R"])) # HUNGARY
print(findWord(["I>F", "W>I", "S>W", "F>T"])) # SWIFT
print(findWord(["R>T", "A>L", "P>O", "O>R", "G>A", "T>U", "U>G"])) # PORTUGAL
print(findWord(["W>I", "R>L", "T>Z", "Z>E", "S>W", "E>R", "L>A", "A>N", "N>D", "I>T"])) # SWITZERLAND
```
